```python
import numpy as np
import random as rand
import matplotlib.pyplot as plt
%matplotlib inline

# Defining function for calculating tau and mu
def function_for_tau_and_mu(a0,a_mu):
    
    # Generating two random numbers
    r = np.random.rand(2)
    r1 = r[0]
    r2 = r[1]
    
    # Calculating tau (i.e time interval for next reaction)
    tau = (1/a0)*np.log(1/r1)
  
    # Calculating mu (i.e reaction number which will ocuur next)
    v = 0
    S = a0*r2 

    while S > 0:
        S = S - a_mu[v]
        v = v + 1
    
    # We have used reaction number staring from 0 to 2
    mu = v - 1
    
    return tau, mu 

```

```python
# Defining function for implementing gillespie method
def gillespie(X,tmax,c):

  # Defining a_mu array
  a_mu = np.zeros(3)

  # Initial time 
  t = 0

  # Count for calculating total number of reactions ocuuring
  count = 0

  # Defining h array
  h = np.zeros(3)

  # Initial population storing in list
  Y1 = [X[0]]
  Y2 = [X[1]]

  # Storing time values in list
  store_t = [t]

  
  while t < tmax:
    # Assigning values to h coefficents
    h[0] = X[0]
    h[1] = X[0]*X[1]
    h[2] = X[1]
    
    # Calculating a_mu and a0 values
    for i in range(3):
      a_mu[i] = h[i]*c[i]
    a0 = sum(a_mu)

    # Calculating tau and mu for a_mu and a0
    tau , mu = function_for_tau_and_mu(a0,a_mu)

    # updating time and count
    t = t + tau
    count = count + 1

    # Breaking loop when both populations becomes equal to or less than equal to zero
    if X[0] <= 0 and X[1] <= 0:
      break


    # Implementing reactions
    if mu == 0:
      X[0] = X[0] + 1
      X[1] = X[1]
    elif mu == 1:
      X[0] = X[0] - 1
      X[1] = X[1] + 1
    else: 
      X[0] = X[0]
      X[1] = X[1] - 1


    # Storing new population value and time 
    Y1.append(X[0])
    Y2.append(X[1])
    store_t.append(t)

  # Converting list to array
  Y1 = np.array(Y1)
  Y2 = np.array(Y2)
  store_t = np.array(store_t)
  return count, Y1, Y2, store_t
```

```python
i = 0
# Simulating three times
for i in range (3):
  
  # Initial population
  X = np.array([1000, 1000])

  # Initial reaction rates
  c = np.array([10,0.01,10])

  # Maximum time for simulation
  tmax = 30

  # Implementing gillespie function for calculating count, Y1, Y2, t
  H = gillespie(X,tmax,c) #(count, Y1, Y2, t)
  print(H)


  # Ploting Y1 vs t
  plt.plot(H[3],H[1],color='green',linewidth=0.5)
  plt.xlabel('t') 
  plt.ylabel('Y1') 
  plt.title("Y1 vs t")
  plt.show()

  # Plotting Y2 vs t
  plt.plot(H[3],H[2],color='red',linewidth=0.5)
  plt.xlabel('t') 
  plt.ylabel('Y2') 
  plt.title("Y2 vs t")
  plt.show()

  # Plotting Y1 vs Y2
  plt.plot(H[1],H[2],linewidth = 0.1,color='black',linestyle='dashed')
  plt.xlabel('Y1') 
  plt.ylabel('Y2') 
  plt.title("Y1 vs Y2")
  plt.show()

  i = i + 1 
```
